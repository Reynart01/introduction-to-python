## Task 2

`[0.2]` Create a new column `whoWon` which contains `"red"` or `"blue"` based on the `blueWins` column.

````python
lol_data['whoWon'] = lol_data['blueWins'].map(lambda b: "red" if b == 0 else "blue")
lol_data.sample(5)
````

`[0.2]` Create a new column `firstKill` which contains `"red"` or `"blue"` based on either `blueFirstBlood` or `redFirstBlood`.

````python
lol_data['firstKill'] = lol_data['blueFirstBlood'].map(lambda b: "blue" if b == 1 else "red")
lol_data.sample(5)
````

`[0.1]` Set the dtype of the `whoWon` and `firstKill` to categorical.

````python
lol_data['whoWon'].astype('category')
lol_data['firstKill'].astype('category')
````

`[0.5]` Cast all `float` columns to `int`, while setting each value to its ceiling.

````python
for (name, values) in lol_data.iteritems():
    if values.dtype == float:
        lol_data[name] = values.map(lambda v: int(v + 1) if int(v) < v else int(v)) 
        
lol_data.sample(5)

        
````

## Task 3

`[1.0]` Create two new columns: `blueFinalWards` and `redFinalWards` that are calculated in the following manner:

```
blueFinalWards = blueWardsPlaced - redWardsDestroyed
redFinalWards = redWardsPlaced - blueWardsDestroyed
```

````python
def get_bfw(row):
    row['blueFinalWards'] = row['blueWardsPlaced'] - row['redWardsDestroyed'] 
    return row

def get_rfw(row):
    row['redFinalWards'] = row['redWardsPlaced'] - row['blueWardsDestroyed'] 
    return row

lol_data = lol_data.apply(get_bfw, axis='columns')
lol_data = lol_data.apply(get_rfw, axis='columns')
````

## Task 4

`[0.5]` Create two new columns: `blueMinionsKilledOutsideJungle` and `redMinionsKilledOutsideJungle` which is a difference of two specific columns in the dataset and save it to the variable `lol_data_updated`.

````python
def get_bminions(row):
    row['blueMinionsKilledOutsideJungle'] = row['blueTotalMinionsKilled'] - row['blueTotalJungleMinionsKilled'] 
    return row

def get_rminions(row):
    row['redMinionsKilledOutsideJungle'] = row['redTotalMinionsKilled'] - row['redTotalJungleMinionsKilled']
    return row

lol_data_updated = lol_data.apply(get_bminions, axis='columns')
lol_data_updated = lol_data_updated.apply(get_rminions, axis='columns')
````

`[0.5]` Remove the following columns: `blueTotalMinionsKilled`, `redTotalMinionsKilled`, `blueTotalJungleMinionsKilled`, `redTotalJungleMinionsKilled` and save this to the previous variable.

````python
lol_data = lol_data_updated.drop(['blueTotalMinionsKilled', 'redTotalMinionsKilled', 'blueTotalJungleMinionsKilled', 'redTotalJungleMinionsKilled'], axis='columns')
````

## Task 5

`[0.8]` Create a new variable `lol_data_group` where you groupby `whoWon` and `firstKill`, with the following aggregations:

- `blueTotalGold` - sum
- `redTotalGold` -  sum
- `blueKills` - mean
- `redKills` - mean
- `blueGoldPerMin` - max
